```python
import os
from time import time
from tqdm import tqdm
import numpy as np
import pandas as pd
import torch
from torch.nn import Linear, ELU, Dropout  # Import ELU activation function
import torch.optim as optim
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torch.utils.data import DataLoader
import torchvision.datasets as datasets

import torchvision
from torchvision.datasets import ImageFolder
from torchvision.models import resnet18
from torchvision.transforms import transforms
import matplotlib.pyplot as plt

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Data Transformer
tfm_train = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.RandomResizedCrop(224),  # Random crop to (224, 224)
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),  # Random color jitter
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
# Additional test-time augmentation
tfm_test = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.CenterCrop(224),  # Center crop to (224, 224)
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])


# Dataset paths
train_ds = datasets.ImageFolder(root="C:/Users/bhagirathi/Downloads/archive (3)/Image", transform=tfm_train)
test_ds = datasets.ImageFolder(root="C:/Users/bhagirathi/Downloads/FIRE/Our_Dataset_610", transform=tfm_test)

# DataLoader
train_loader = DataLoader(train_ds, batch_size=64, shuffle=True)
test_loader = DataLoader(test_ds, batch_size=64, shuffle=False)

# Model
model = resnet18(pretrained=True)

# Replace Output of Fully Connected Layer with Number of Labels for our Classification Problem
model.fc = torch.nn.Sequential(
    Linear(512, 256),
    ELU(),  # Change activation function from ReLU to ELU
    Dropout(p=0.7),
    Linear(256, 2)
)

model = model.to(device)

# Optimiser
optimizer = optim.Adam(model.parameters(), lr=0.0001, weight_decay=0.0001)  # Adjusted learning rate

# Define the ReduceLROnPlateau scheduler
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5, verbose=True, threshold=0.0001, threshold_mode='rel', cooldown=0, min_lr=0, eps=1e-08)

# Loss Function
loss_fn = torch.nn.CrossEntropyLoss()

train_acc_list = []
test_acc_list = []
train_losses = []
test_losses = []

# epochs
num_epochs = 25

for epoch in range(num_epochs):
    start = time()

    train_correct = 0
    test_correct = 0
    train_loss = 0.0
    test_loss = 0.0

    # Train
    model.train()

    for xtrain, ytrain in tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs}', unit="batch"):
        optimizer.zero_grad()

        xtrain = xtrain.to(device)
        ytrain = ytrain.to(device)

        train_prob = model(xtrain)

        loss = loss_fn(train_prob, ytrain)
        loss.backward()
        optimizer.step()

        train_pred = torch.argmax(train_prob, axis=1)
        train_correct += torch.sum(train_pred == ytrain).item()
        train_loss += loss.item() * xtrain.size(0)

    train_loss = train_loss / len(train_ds)
    train_losses.append(train_loss)
    train_acc = train_correct / len(train_ds)
    train_acc_list.append(train_acc)

    # Evaluate
    model.eval()

    for xtest, ytest in test_loader:
        xtest = xtest.to(device)
        ytest = ytest.to(device)

        test_prob = model(xtest)
        test_pred = torch.argmax(test_prob, axis=1)
        test_correct += torch.sum(test_pred == ytest).item()

        loss = loss_fn(test_prob, ytest)
        test_loss += loss.item() * xtest.size(0)

    test_loss = test_loss / len(test_ds)
    test_losses.append(test_loss)
    test_acc = test_correct / len(test_ds)
    test_acc_list.append(test_acc)

    scheduler.step(test_acc)

    end = time()
    duration = (end - start) / 60

    print(f"Epoch: {epoch+1}/{num_epochs}, Time: {duration:.2f} mins, Train Acc: {train_acc:.4f}, Test Acc: {test_acc:.4f}")

# Save results to CSV
data_accuracy = {
    'Epoch': range(1, num_epochs + 1),
    'Train Accuracy': train_acc_list,
    'Test Accuracy': test_acc_list
}

data_loss = {
    'Epoch': range(1, num_epochs + 1),
    'Train Loss': train_losses,
    'Test Loss': test_losses
}

df_accuracy = pd.DataFrame(data_accuracy)
df_loss = pd.DataFrame(data_loss)

df_accuracy.to_csv('C:/Users/bhagirathi/Downloads/FIRE/resnet_EAR_accuracy.csv', index=False)
df_loss.to_csv('C:/Users/bhagirathi/Downloads/FIRE/resnet_EAR_accuracy.csv', index=False)

# Calculate average accuracies and losses
avg_train_accuracy = sum(train_acc_list) / len(train_acc_list)
avg_test_accuracy = sum(test_acc_list) / len(test_acc_list)
avg_train_loss = sum(train_losses) / len(train_losses)
avg_test_loss = sum(test_losses) / len(test_losses)

# Print average accuracies and losses
print("Average Train Accuracy:", avg_train_accuracy)
print("Average Test Accuracy:", avg_test_accuracy)
print("Average Train Loss:", avg_train_loss)
print("Average Test Loss:", avg_test_loss)

# Plotting
plt.plot(range(1, num_epochs+1), train_acc_list, marker='o', label='Train')
plt.plot(range(1, num_epochs+1), test_acc_list, marker='o', label='Test')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training and Test Accuracy vs Epochs')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(range(1, num_epochs+1), train_losses, label='Training Loss')
plt.plot(range(1, num_epochs+1), test_losses, label='Testing Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Testing Loss vs Epochs')
plt.legend()
plt.show()
```

```python
import random

# Set model to evaluation mode
model.eval()

# Function to convert class index to class label
def get_class_label(class_idx):
    return train_ds.classes[class_idx]

# Select random 5 images from the test dataset
random_indices = random.sample(range(len(test_ds)), 10)

for idx in random_indices:
    image, label = test_ds[idx]

    # Add batch dimension and move to device
    image = image.unsqueeze(0).to(device)

    # Forward pass
    with torch.no_grad():
        output = model(image)

    # Get predicted class
    pred_idx = torch.argmax(output, dim=1).item()
    pred_label = get_class_label(pred_idx)

    # Get ground truth label
    true_label = get_class_label(label)

    # Display the result
    plt.imshow(image.squeeze().cpu().permute(1, 2, 0))
    plt.axis('off')
    
    # Print predicted label and ground truth label
    print(f'Prediction: {pred_label}, True Label: {true_label}')
    
    # Determine if it's fire or non-fire
    if pred_label == "Fire":
        print("Fire Detected")
    else:
        print("Non-Fire Detected")
    
    plt.show()
```
