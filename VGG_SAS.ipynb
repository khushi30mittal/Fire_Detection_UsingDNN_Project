```python
import pandas as pd
import matplotlib.pyplot as plt

# Import data from Notebook 1
data1 = pd.read_csv('/content/drive/MyDrive/CSVs/VGG_accuarcy.csv')

# Import data from Notebook 2
data2 = pd.read_csv('/content/drive/MyDrive/resnet_PresentSAS_accuracy.csv')

# Merge data from both notebooks based on a common column (e.g., 'Epoch')
merged_data = pd.merge(data1, data2, on='Epoch', suffixes=('_VGG', '_SAS'))

# Ensure all epochs from 1 to 25 are present
all_epochs = pd.DataFrame({'Epoch': range(1, 26)})
merged_data = pd.merge(all_epochs, merged_data, on='Epoch', how='left')

# Increase the size of the graph
plt.figure(figsize=(12, 9))

# Plotting
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy_VGG'], "ro--", label='VGG-19')
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy_SAS'], "bo--", label='SAS')

plt.xlabel('Epoch')
plt.ylabel('Test Accuracy')
plt.title('Comparison Between Test Accuracies of VGG-19 and SAS')
plt.legend(loc='upper left')

# Set x-axis ticks to display all epochs from 1 to 25
plt.xticks(range(1, 26))
plt.grid(True)

plt.show()
```
