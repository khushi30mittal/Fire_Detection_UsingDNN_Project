```python
import pandas as pd
import matplotlib.pyplot as plt

# Import data from Notebook 1
data1 = pd.read_csv('/content/drive/MyDrive/csvs/resnet_SAR_accuracy.csv')

# Import data from Notebook 2
data2 = pd.read_csv('/content/drive/MyDrive/csvs/resnet_SAS_accuracy.csv')
data3 = pd.read_csv('/content/drive/MyDrive/csvs/resnet_SSR_accuracy.csv')
data4 = pd.read_csv('/content/drive/MyDrive/csvs/resnet_SSS_accuracy.csv')

# Merge data from both notebooks based on a common column (e.g., 'Epoch')
merged_data = pd.merge(data1, data2, on='Epoch', suffixes=('_SAR', '_SAS'))
merged_data = pd.merge(merged_data, data3, on='Epoch', suffixes=('', '_SSR'))
merged_data = pd.merge(merged_data, data4, on='Epoch', suffixes=('', '_SSS'))
# Ensure all epochs from 1 to 25 are present
all_epochs = pd.DataFrame({'Epoch': range(1, 26)})
merged_data = pd.merge(all_epochs, merged_data, on='Epoch', how='left')



# Increase the size of the graph
plt.figure(figsize=(10, 6))  # Width: 10 inches, Height: 6 inches

# Plotting
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy_SAR'], color='blue', marker='o', label='SAR')
plt.legend(loc='upper left')
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy_SAS'], color='green', marker='^', label='SAS')
plt.legend(loc='upper left')
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy'], color='red', marker='s', label='SSR')
plt.legend(loc='upper left')
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy_SSS'], color='orange', marker='*', label='SSS')
plt.legend(loc='upper left')

plt.xlabel('Epoch')
plt.ylabel('Test_Accuracy')
plt.title('Comparison of SAR, SAS, SSR, and SSS')
# Set x-axis ticks to display all epochs from 1 to 25
plt.xticks(range(1, 26))
plt.grid(True)

plt.show()
```
