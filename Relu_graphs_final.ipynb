```python
import pandas as pd
import matplotlib.pyplot as plt

# Import data from Notebook 1
data1 = pd.read_csv('/content/drive/MyDrive/csvs/resnet_RAR_accuracy1.csv')

# Import data from Notebook 2
data2 = pd.read_csv('/content/drive/MyDrive/csvs/resnet_RAS_accuracy.csv')
data3 = pd.read_csv('/content/drive/MyDrive/csvs/resnet_RSR_accuracy.csv')
data4 = pd.read_csv('/content/drive/MyDrive/csvs/resnet_RSS_accuracy.csv')

# Merge data from both notebooks based on a common column (e.g., 'Epoch')
merged_data = pd.merge(data1, data2, on='Epoch', suffixes=('_RAR', '_RAS'))
merged_data = pd.merge(merged_data, data3, on='Epoch', suffixes=('', '_RSR'))
merged_data = pd.merge(merged_data, data4, on='Epoch', suffixes=('', '_RSS'))
# Ensure all epochs from 1 to 25 are present
all_epochs = pd.DataFrame({'Epoch': range(1, 26)})
merged_data = pd.merge(all_epochs, merged_data, on='Epoch', how='left')



# Increase the size of the graph
plt.figure(figsize=(10, 6))  # Width: 10 inches, Height: 6 inches

# Plotting
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy_RAR'], color='blue', marker='o', label='RAR')
plt.legend(loc='upper left')
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy_RAS'], color='green', marker='^', label='RAS')
plt.legend(loc='upper left')
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy'], color='red', marker='s', label='RSR')
plt.legend(loc='upper left')
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy_RSS'], color='orange', marker='*', label='RSS')
plt.legend(loc='upper left')

plt.xlabel('Epoch')
plt.ylabel('Test_Accuracy')
plt.title('Comparison of RAR, RAS, RSR, and RSS')
# Set x-axis ticks to display all epochs from 1 to 25
plt.xticks(range(1, 26))
plt.grid(True)

plt.show()
```
