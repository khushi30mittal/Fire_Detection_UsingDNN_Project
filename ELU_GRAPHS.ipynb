```python
import pandas as pd
import matplotlib.pyplot as plt

# Import data from Notebook 1
data1 = pd.read_csv('/content/drive/MyDrive/CSVs/resnet_EAR_Accuracy.csv')

# Import data from Notebook 2
data2 = pd.read_csv('/content/drive/MyDrive/CSVs/resnet_EAS_accuracy.csv')
data3 = pd.read_csv('/content/drive/MyDrive/CSVs/resnet_ESR_accuracy.csv')
data4 = pd.read_csv('/content/drive/MyDrive/CSVs/resnet_ESS_accuracy.csv')

# Merge data from both notebooks based on a common column (e.g., 'Epoch')
merged_data = pd.merge(data1, data2, on='Epoch', suffixes=('_EAR', '_EAS'))
merged_data = pd.merge(merged_data, data3, on='Epoch', suffixes=('', '_ESR'))
merged_data = pd.merge(merged_data, data4, on='Epoch', suffixes=('', '_ESS'))

# Ensure all epochs from 1 to 25 are present
all_epochs = pd.DataFrame({'Epoch': range(1, 26)})
merged_data = pd.merge(all_epochs, merged_data, on='Epoch', how='left')

# Increase the size of the graph
plt.figure(figsize=(12, 9))  # Width: 10 inches, Height: 6 inches

# Plotting
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy_EAR'], "ro--" , label='EAS')
plt.legend(loc='upper left')
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy_EAS'], "bo:", label='ESR')
plt.legend(loc='upper left')
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy'], "go--", label='ESS')
plt.legend(loc='upper left')
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy_ESS'], "ko:", label='RSS')
plt.legend(loc='upper left')

plt.xlabel('Epoch')
plt.ylabel('Test Accuracy')
plt.title('Comparison Between Test Accuracies of EAR ,EAS , ESR, and ESS')
# Set x-axis ticks to display all epochs from 1 to 25
plt.xticks(range(1, 26))
plt.grid(True)
plt.show()
```
