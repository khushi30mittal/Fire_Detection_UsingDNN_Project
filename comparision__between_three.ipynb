```python
import pandas as pd
import matplotlib.pyplot as plt

# Import data from Notebook 1
data1 = pd.read_csv('/content/drive/MyDrive/csvs/resnet_RAR_accuracy1.csv')

# Import data from Notebook 2
data2 = pd.read_csv('/content/drive/MyDrive/resnet_EAR_Accuracy.csv')
data3 = pd.read_csv('/content/drive/MyDrive/csvs/resnet_SAR_accuracy.csv')

# Merge data from both notebooks based on a common column (e.g., 'Epoch')
merged_data = pd.merge(data1, data2, on='Epoch', suffixes=('_RAR', '_EAR'))
merged_data = pd.merge(merged_data, data3, on='Epoch', suffixes=('', '_SAR'))

# Ensure all epochs from 1 to 25 are present
all_epochs = pd.DataFrame({'Epoch': range(1, 26)})
merged_data = pd.merge(all_epochs, merged_data, on='Epoch', how='left')

# Increase the size of the graph
plt.figure(figsize=(12, 9))  # Width: 10 inches, Height: 6 inches

# Plotting
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy_RAR'], "ro--" , label='RAR')
plt.legend(loc='upper left')
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy_EAR'], "bo:", label='EAR')
plt.legend(loc='upper left')
plt.plot(merged_data['Epoch'], merged_data['Test Accuracy'], "mo:", label='SAR')
plt.legend(loc='upper left')

plt.xlabel('Epoch')
plt.ylabel('Test Accuracy')
plt.title('Comparison Between Test Accuracies of RAR , EAR AND SAR')
# Set x-axis ticks to display all epochs from 1 to 25
plt.xticks(range(1, 26))
plt.grid(True)
plt.show()
```
